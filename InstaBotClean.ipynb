```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
driver = webdriver.Chrome(executable_path='/Users/hardikdudeja/Downloads/chromedriver')
```

```python
def login(username, password):
    driver.get('https://www.instagram.com/')
    # waiting till the element appears
    btn_login = WebDriverWait(driver, 10).until( 
            EC.presence_of_element_located((By.XPATH, '/html/body//article/div[2]/div/div/form/div/div[3]')) 
        )
    nameField = driver.find_element_by_name('username')
    passwordField = driver.find_element_by_name('password')
    nameField.send_keys(username)
    passwordField.send_keys(password)     
    btn_login.click()
```

```python
def loginProcess():
    try:
        btn_notNow = driver.find_element_by_xpath('//button[contains(@class, "yWX7d")]')
        btn_notNow.click()
    except:
        pass
    time.sleep(3)
    try:
        btn_no = driver.find_element_by_xpath('//button[contains(@class, "HoLwm")]')
        btn_no.click()
    except:
        pass
```

```python
#Login process
login('sample', '**********I')
time.sleep(3)
loginProcess()
```

```python
def search(string):
    searchBox = driver.find_element_by_xpath('//input[contains(@class, "XTCLo")]')
    searchBox.clear()
    searchBox.send_keys(string)
    #using explicit wait
    results = WebDriverWait(driver, 10).until( 
                EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "fuqBx")]//a')) 
            )
    li_names = []
    for ele in results:
        ele = ele.get_attribute('href')
        ele = ele.split('/')
        if(len(ele) == 5):
            if(ele[-2] == 'food'):
                continue
            li_names.append(ele[-2])
    return li_names
```

```python
#Q2
li_usernames = search('food')
for ele in li_usernames:
    print(ele)
```

```python
#this function opens the profile with the given username
def searchAndOpenProfile(string):
    searchBox = driver.find_element_by_xpath('//input[contains(@class, "XTCLo")]')
    searchBox.clear()
    searchBox.send_keys(string)
    time.sleep(1)
    result = WebDriverWait(driver, 10).until( 
                EC.presence_of_element_located((By.XPATH, '//div[contains(@class, "fuqBx")]/div/a')) 
            )
    result.click()
    
```

```python
#following the page
def follow(username):
    searchAndOpenProfile(username)
    btnFollow = driver.find_element_by_xpath('//span[contains(@class,"vBF20")]')
    btnFollow.click()
    time.sleep(2)
    try:
        unfollowWindow = driver.find_element_by_class_name('piCib')
        print('Already following the Page')
        time.sleep(1)
        btnCancel = driver.find_element_by_xpath('//button[contains(@class, "HoLwm")]')
        btnCancel.click()
    except:
        print('following the page now')
    
```

```python
def unfollow(username):
    searchAndOpenProfile(username)
    btnFollow = driver.find_element_by_xpath('//span[contains(@class,"vBF20")]')
    btnFollow.click()
    time.sleep(2)
    try:
        unfollowWindow = driver.find_element_by_class_name('piCib')
        time.sleep(1)
        btnUnfollow = driver.find_element_by_xpath('//button[contains(@class, "-Cab_")]')
        btnUnfollow.click()
        print('Unfollowed')
    except:
        print('Page was already unfollowed')
        btnFollow.click()
        time.sleep(1)
        unfollowWindow = driver.find_element_by_class_name('piCib')
        time.sleep(1)
        btnUnfollow = driver.find_element_by_xpath('//button[contains(@class, "-Cab_")]')
        btnUnfollow.click()
        
    
```

```python
#only when profile is open
def openFirstPost():
    try:
        firstPost = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'eLAPa'))
            )
        firstPost.click()
    except:
        pass
```

```python
def likeAndInform():
    likeBtn = driver.find_element_by_xpath('//span[@class = "fr66n"]/button')
    svg = driver.find_element_by_xpath('//span[@class = "fr66n"]/button/div/span/*[name()="svg"]')
#     print(svg.get_attribute('aria-label'))
    
    if svg.get_attribute('aria-label') == 'Unlike':
        print('Post already liked')
    else:
        likeBtn.click()
        print('Liked the post now')
    
```

```python
def dislikeAndInform():
    likeBtn = driver.find_element_by_xpath('//span[@class = "fr66n"]/button')
    svg = driver.find_element_by_xpath('//span[@class = "fr66n"]/button/div/span/*[name()="svg"]')
    
    if svg.get_attribute('aria-label') == 'Like':
        print('Post was not liked')
    else:
        likeBtn.click()
        print('Disliked the post now')
    
```

```python
def nextPost():
    btn_next = driver.find_element_by_xpath('//div[contains(@class, "DdSX2")]/a[last()]')
    btn_next.click()
    time.sleep(1)
```

```python
#this is done using the next post button
def likingSomePosts(username, n):
    searchAndOpenProfile(username)
    time.sleep(1)
    openFirstPost()
    time.sleep(1)
    for i in range(n-1):
        print('Post', i+1, end=' : ')
        likeAndInform()
        time.sleep(2)
        nextPost()
        time.sleep(1)
    print('Post', n, end=' : ')
    likeAndInform()
    btnClose = driver.find_elements_by_xpath('//button[contains(@class, "wpO6b")]')[-1]
    btnClose.click()
```

```python
likingSomePosts('dilsefoodie', 30)
```

```python
def dislikingSomePosts(username, n):
    searchAndOpenProfile(username)
    time.sleep(1)
    openFirstPost()
    time.sleep(1)
    for i in range(n-1):
        print('Post', i+1, end=' : ')
        dislikeAndInform()
        time.sleep(2)
        nextPost()
        time.sleep(1)
    print('Post', n, end=' : ')
    dislikeAndInform()
    
    btnClose = driver.find_elements_by_xpath('//button[contains(@class, "wpO6b")]')[-1]
    btnClose.click()
```

```python
def openFollowersList(username, n):
    searchAndOpenProfile(username)
    time.sleep(2)
    followerList = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "k9GMp")]/li[2]/a'))
    )
    followerList.click()
    scroll()
    time.sleep(2)
    return getFollowers(n)
```

```python
def scroll():
    for i in range(50):
        current_height= driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0, arguments[0]);',current_height) 
        time.sleep(1)
```

```python
def getFollowers(n):
    li = driver.find_element_by_class_name('PZuss')
    i=0
    titles = driver.execute_script('return document.getElementsByClassName("FPmhX");')
#     titles = li.find_elements_by_xpath('//span[contains(@class, "Jv7Aj mArmR MqpiF  ")]/a')
    li_followers = []
    for ele in titles:
        li_followers.append(ele.text)
        i += 1
        if i == n:
            break
    btnClose = driver.find_elements_by_class_name('WaOAr')[-1].find_element_by_xpath('button[contains(@class, "wpO6b")]').click()
    return li_followers
        
```

```python
def goToProfile():
    optionsPanel = driver.find_element_by_class_name('_47KiJ')
    accountButton = optionsPanel.find_elements_by_class_name('Fifk5')[4]
    accountButton.click()
    time.sleep(1)
    popupWindow = accountButton.find_element_by_class_name('_01UL2')
    profileBtn = popupWindow.find_element_by_class_name('-qQT3')
    profileBtn.click()
    
```

```python
def getMyFollowerList():
    goToProfile()
    time.sleep(2)
    followerBtn = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "k9GMp")]/li[2]/a'))
    )
    followerBtn.click()
    li_followers = []
    try:
        scroll()
        li = driver.find_element_by_class_name('PZuss')
        i=0
        titles = li.find_elements_by_xpath('//span[contains(@class, "Jv7Aj mArmR MqpiF  ")]/a')
        for ele in titles:
            li_followers.append(ele.get_attribute('title'))
            i += 1

    except:
        print('0 followers')
        
    topPane = driver.find_element_by_class_name('eiUFA')
    exitBtn = topPane.find_elements_by_class_name('WaOAr')[1]
    exitBtn.click()
    return li_followers
    
    
```

```python
def getMyFollowing():
    goToProfile()
    followingBtn = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "k9GMp")]/li[3]/a'))
    )
    followingBtn.click()
    li_following = []
    try:
        scroll()
        li = driver.find_element_by_class_name('PZuss')
        i=0
        titles = li.find_elements_by_xpath('//span[contains(@class, "Jv7Aj mArmR MqpiF  ")]/a')
        for ele in titles:
            li_following.append(ele.get_attribute('title'))
            i += 1
    except:
        print('0 following')
    topPane = driver.find_element_by_class_name('eiUFA')
    exitBtn = topPane.find_elements_by_class_name('WaOAr')[1]
    exitBtn.click()
    return li_following
    
```

```python
def watchStory(username):
    searchAndOpenProfile(username)
    try:
        story = driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]')
        canvas = driver.find_element_by_class_name('CfWVH')
        height = int(canvas.get_attribute('height'))
        if height == 137 or height == 252:
            st.click()
            print('taking a look at the story right now')
        else:
            print('already seen the story')
    except:
        print('No story')
```
